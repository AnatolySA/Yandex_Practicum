```python
def filling(data, data_agg, column):
    """
    Функция принимает исходный датафрейм с пропущенными
    значениями, вспомогательный датафрейм, значениями 
    которого будут заполнены пропуски, название ряда,
    который необходимо заполнить.
    """
    models = data['model'].unique().tolist()
    for model in models:
        data.loc[data['model'] == model, column] = np.where(
        data.loc[data['model'] == model, column].isna() == True, 
        data_agg.loc[data_agg['model'] == model, column],
        data.loc[data['model'] == model, column])
```

Для заполнения пропусков в признаке `repaired` будем сортировать объявления по медианной цене. Для этого создадим сводную таблицу, в которую для каждой модели передадим значения медианной цены в целом, а также в разбивке по нахождению автомобиля в ремонте.

```python
# Создаем вспомогательную таблицу с медианной ценой
# в зависимости от нахождения автомобиля в ремонте
train_repaired_pivot = train.pivot_table(values='price',
                                         index=['model'],
                                         columns=['repaired'],
                                         aggfunc='median').reset_index()

# Создаем вспомогательную таблицу с медианной ценой
# для различных моделей автомобилей в целом
train_price = train.pivot_table(values='price', 
                                index=['model'], 
                                aggfunc='median').reset_index()

# Объединяем две вспомогательные таблицы в одну
train_price = train_price.merge(train_repaired_pivot, 
                                on='model', 
                                how='left')

# Заполням пропущенные значения "заглушкой"
train['repaired'] = train['repaired'].fillna('replace')
```

```python
# Формируем список моделей автомобилей из объявлений
models = train['model'].unique().tolist()
# Задаем константы для определения уровня цены
NO_LEVEL = 1.2
YES_LEVEL = 0.8
# Создаем цикл, 
for i in range(len(models)):
    model = models[i]
    # Передаем значение медианной цены для модели i
    median_price = train_price.loc[
        train_price['model'] == model, 'price']
    # В случае отклонения цены в объявлении от медианной цены
    # для модели в целом более чем на 20% вниз (YES_LEVEL),
    # будем считать, что автомобиль ремонтировался
    train.loc[((train['model'] == model) &
               (train['price'] < YES_LEVEL * int(median_price)) &
               (train['repaired'] == 'replace')),
              'repaired'] = 'yes'
    # В случае отклонения цены в объявлении от медианной цены
    # для модели в целом более чем на 20% вверх (NO_LEVEL),
    # будем считать, что автомобиль не ремонтировался
    train.loc[((train['model'] == model) &
               (train['price'] > NO_LEVEL * int(median_price)) &
               (train['repaired'] == 'replace')),
              'repaired'] = 'no'
```

#### Аномальные значения

```python
LOW_PRICE = 0.05
HIGHT_PRICE = 2.5
for i in range(len(models)):
    model = models[i]
    # Передаем значение медианной цены для модели i    
    median_price = train_price.loc[
        train_price['model'] == model, 'price']
    # В случае отклонения цены в объявлении от медианной цены
    # для модели в целом более чем на 95% вниз (LOW_PRICE),
    # будем считать значение неадекватным    
    train.loc[((train['model'] == model) &
               (train['price'] < LOW_PRICE * int(median_price)),
              'price')] = 0
    # В случае отклонения цены в объявлении от медианной цены
    # для модели в целом более чем на 150% вверх (HIGHT_PRICE),
    # будем считать значение неадекватным 
    train.loc[((train['model'] == model) &
               (train['price'] > HIGHT_PRICE * int(median_price)),
              'price')] = 0
```

```python
train = train.query('model != "samara"').copy()


train_numer_pivot = train.pivot_table(index=['model'],
                                      values=['power', 'registration_year'],
                                      aggfunc='median').reset_index()
```

```python
models = train['model'].unique().tolist()
# Задаем константы для определения уровня мощности и года
POWER_LEVEL = 0.5
YEAR_LEVEL = 0.9
YEAR = 2016
# Создаем цикл, 
for i in range(len(models)):
    model = models[i]
    # Передаем значение медианной мощности и года 
    # регистрации для модели i
    median_power = train_numer_pivot.loc[
        train_numer_pivot['model'] == model, 'power']
    median_year = train_numer_pivot.loc[
        train_numer_pivot['model'] == model, 'registration_year']
    # В случае отклонения мощности автомобиля в объявлении от 
    # медианного значения более чем на 50% в любую сторону (POWER_LEVEL)
    # заменим значение на медианное
    train.loc[((train['model'] == model) &
               ((train['power'] < POWER_LEVEL * int(median_power)) |
                (train['power'] > (1 + POWER_LEVEL) * int(median_power))
               )
              ),
              'power'] = float(median_power)
    # В случае отклонения мощности автомобиля в объявлении от 
    # медианного значения более чем на 90% в любую сторону (YEAR_LEVEL)
    # или превышении значения текущего года заменим значение на медианное
    train.loc[((train['model'] == model) &
               ((train['registration_year'] < YEAR_LEVEL * int(median_year)) |
                (train['registration_year'] > (1 + YEAR_LEVEL) * int(median_year)) |
                (train['registration_year'] > YEAR)
               )
              ),
              'registration_year'] = float(median_year)
```

### Подготовка признаков для обучения моделей

Обратим более детальное внимание на корреляционную матрицу. Легко заметить, что тип кузова, бренд и модель автомобиля имеют высокую попарную корреляцию. Во избежание мультиколлинеарности признаков оставим только один из них. В качестве объясняющего признака выберем модель автомобиля, что, на наш взгляд, является достаточно логичным:
 - бренд автомобиля является более широким понятием, так как может содержать в себе разные модели;
 - тип кузова может быть одинаковым у автомобилей разных ценовых сегментов;
 - модельный ряд автомобилей представлен, как правило, небольшим разнообразием с точки зрения типа кузова, при этом влияние на стоимость такое различие оказывает минимальное.

```python
# Определим определяющие (features) и определяемые (target) признаки для каждой выборки:
# Тренировочная выборка:
features_train = train.drop(['date_crawled',
                             'price',
                             'vehicle_type',
                             'registration_month',
                             'brand',
                             'date_created',
                             'last_seen'], 
                            axis = 1)
target_train = train['price']
# Тестовая выборка:
features_test = test.drop(['date_crawled', 
                           'price',
                           'vehicle_type',
                           'registration_month',
                           'brand',
                           'date_created',
                           'last_seen'], 
                           axis = 1)
target_test = test['price']
```

Проведем энкодинг категориальных признаков.

```python

encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

encoding_columns = ['registration_year', 
                    'gearbox', 
                    'power', 
                    'model', 
                    'fuel_type', 
                    'repaired']


features_train[encoding_columns] = encoder.fit_transform(features_train[encoding_columns])

for column in features_train.columns:
    features_train[column] = features_train[column].astype(int)
```

```python
STATE = 442
```

### RandomForestRegressor

```python
%%time
# Определяем модель
model = RandomForestRegressor(random_state=STATE)

# Задаем параметры для GridSearchCV
# params = {'max_depth':range(10, 26, 5), 
#           'n_estimators': range(100, 501, 100),
#           'min_samples_leaf': range(1, 8, 3)
#                  }

params = {
    'max_depth': [20],
    'min_samples_leaf': [1], 
    'n_estimators': [500]
}

grid_rand_forest = GridSearchCV(model, 
                                params,
                                cv = 3, 
                                scoring = 'neg_mean_squared_error', 
                                n_jobs = -1)
grid_rand_forest.fit(features_train, target_train)

print(f'RMSE для лучшей модели случайного леса составило: '
      f'{-grid_rand_forest.best_score_:.6}')
print(f'Гиперпараметры лучшей модели: {grid_rand_forest.best_params_}')
```

### LightGBM

```python
%%time
# Определяем модель
model = lgb.LGBMRegressor(random_state = STATE)
# Задаем параметры для GridSearchCV
# params = {
#     'max_depth':range(10, 51, 5),
#     'reg_alpha':[0.1, 0.3, 0.5, 0.7],
#     'num_leaves':range(2, 10, 3)}

params = {
    'max_depth': [10],
    'num_leaves': [8], 
    'reg_alpha': [0.1]
}

grid_lgbm = GridSearchCV(model, 
                         params, 
                         cv = 3, 
                         scoring = 'neg_mean_squared_error', 
                         n_jobs = -1)
grid_lgbm.fit(features_train, target_train)

print(f'RMSE для лучшей модели бустинга составило: '
      f'{-grid_lgbm.best_score_:.6}')
print(f'Гиперпараметры лучшей модели: {grid_lgbm.best_params_}')
```

### CatBoostRegressor

```python
%%time
# Определяем модель
cat = CatBoostRegressor(random_state = STATE)
# Задаем параметры для GridSearchCV
# params = {'depth': range(10, 15),
#           'l2_leaf_reg': [3, 5]}

params = {'depth': [14], 
          'l2_leaf_reg': [3]}


grid_cat = GridSearchCV(cat, 
                        params, 
                        cv = 3, 
                        scoring = 'neg_mean_squared_error', 
                        n_jobs = -1, 
                        error_score='raise')
grid_cat.fit(features_train, target_train)

print(f'RMSE для лучшей модели CatBoost: '
      f'{-grid_cat.best_score_:.6}')
print(f'Гиперпараметры лучшей модели: {grid_cat.best_params_}')
```
