## Подготовка данных

```python
# Импортируем библиотеки и модули, которые понадобятся нам 
# для решения поставленной задачи
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
from sklearn.dummy import DummyRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import make_scorer, mean_absolute_error
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
```

```python
# Определим каждый датафрейм
data_full = pd.read_csv('gold_industry_full.csv')
data_train = pd.read_csv('gold_industry_train.csv')
data_test = pd.read_csv('gold_industry_test.csv')
```

### Расчет эффективности обогащения

```python
C = data_train['rougher.output.concentrate_au']
F = data_train['rougher.input.feed_au']
T = data_train['rougher.output.tail_au']

data_train['calc'] = (C * (F - T)) / (F * (C - T)) * 100
mae = mean_absolute_error(data_train['rougher.output.recovery'], data_train['calc'])
print(f'Средняя абсолютная ошибка (MAE) между реальным и расчетным '
      f'значениями эффективности обогащения на этапе \nфлотации '
      f'составила {mae:.2}.')
```

Cредняя абсолютная ошибка — это мера ошибок между парными наблюдениями, выражающими одно и то же явление, например, при сравнении одного метода измерения и альтернативного метода измерения. Значение, находящееся вблизи 0, позволяет сделать вывод о корректном расчете показателя в условиях реального производства. В базу данных переданы корректные значения показателя.

Определим, какие еще признаки отсутствуют в тестовой выборке.

```python
# Определяем пустой список, в который передадим название признаков,
# которых нет в тестовой выборке, но есть в тренировочной
list_diff = []
list_test = data_test.columns
list_train = data_train.columns

for a in list_train:
    if a not in list_test:
        list_diff.append(a)

# Определяем списки, между которыми на основании типа параметра
# распределим отсутствующие признаки

list_calc = []
list_input = []
list_other = []
list_output = []
list_state = []

for element in list_diff:
    if '.input.' in element:
        list_input.append(element)
    elif '.output.' in element:
        list_output.append(element)
    elif '.state.' in element:
        list_state.append(element)
    elif '.calculation.' in element:
        list_calc.append(element)
    else:
        list_other.append(element)

print(f'Признаки, недоступные в тестовой выборке:')        
print(f'характеризующие параметры сырья: {len(list_input)} шт.;')
print(f'характеризующие параметры продукта: {len(list_output)} шт.;')
print(f'характеризующие текущее состояние этапа: {len(list_state)} шт.;')
print(f'являющиеся расчетным характеристиками: {len(list_calc)} шт.;')
print(f'иные: {len(list_other)} шт.')
```

В тестовой выборке отсутствуют признаки, которые неизвестны во время технологического процесса и их возможно измерить только по его завершении; часть признаков, которые вычисляются на основе этих измерений; расчетный признак, добавленный нами на этапе `1.1`. 

```python
# Удалим из тренировочной выборки признаки, 
# которые не понадобятся нам при обучении модели
list_all = list_calc+list_input+list_other+list_output+list_state
data_train_model = data_train.drop(columns = list_all)
data_train_analys = data_train
```

## Анализ данных

### Изменение концентрации металлов (Au, Ag, Pb) на различных этапах обработки

По мере прохождения сырьем, содержащим драгоценные металлы, их концентрация меняется. Проверим концентрацию золота, серебра и свинца на различных этапах обработки.

```python
def conc_change(data, me):
    """
    В функцию передается анализируемый датафрейм и название металла 
    (как химического элемента) в виде строки. 
    Функция возвращает графики распределения концентрации металла 
    на различных этапах обработки.
    """
    # Формируем название признаков, которые будем анализировать
    roug_in = 'rougher.input.feed_' + me.lower()
    roug_out = 'rougher.output.concentrate_' + me.lower()
    prim_clean = 'primary_cleaner.output.concentrate_' + me.lower()
    fin_out = 'final.output.concentrate_' + me.lower()
    
    # Дополнительные преобразования 
    # для корректного вывода названия графика
    name = None
    if me.lower() == 'au':
        name = 'золота'
    elif me.lower() == 'ag':
        name = 'серебра'
    elif me.lower() == 'pb':
        name = 'свинца'
    
    # Формируем график
    sns.histplot(data[roug_in], alpha=0.4, element="step", 
                 label='Исходное сырье', 
                 color='yellow', edgecolor='red')
    sns.histplot(data[roug_out], alpha=0.4,  element="step", 
                 label='Черновой концентрат', 
                 color='orange', edgecolor='red')
    sns.histplot(data[prim_clean], alpha=0.4,  element="step",
                 label='Первчиная очистка',
                 color='blue', edgecolor='blue')
    sns.histplot(data[fin_out], alpha=0.4,  element="step",
                 label='Финальный концентрат',
                 color='green', edgecolor='green')
    plt.title(f'Распределение концентрации {name} на различных этапах очистки', size= 16 , weight= 100)
    plt.xlabel('')
    plt.ylabel('')
    plt.legend(bbox_to_anchor=(0.5, -0.05), loc='upper center', ncol=4)
    plt.show()
```

```python
conc_change(data_train_analys, 'Au')
```

От исходного сырья (золотосодержащей руды) до итогового продукта переработки концентрация золота увеличивается последовательно на каждом этапе. При этом при переходе исходного сырья на стадию чернового концентрата разброс значений сохраняется примерно на том же уровне, хотя средняя концентрация золота увеличивается. На этапе первчиной очистки увеличивается не только концентрация золота, но и разброс относительно среднего значения. Наконец, разброс значений финального концентрата становится меньше, а концентрация золота достигает максимальных значений.  

```python
conc_change(data_train_analys, 'ag')
```

Концентрация серебра в черновом концентрате максимальна. На этапе первичной очистки она снижается, финальный концентрат содержит наименьшую из всех этапов концентрацию серебра, при этом разброс значений относительно среднего также минимальный.

```python
conc_change(data_train_analys, 'PB')
```

Изменение концентрации свинца схоже с изменением концентрации золота. Однако на этапе первичной обработки концентрация свинца максимальная, ее разброс относительно среднего значения достаточно велик. На финальном этапе разброс значений снижается, средняя концентрация немного уменьшается.

## Модель

### Функция для вычисления итогового sMAPE

```python
# Определим константу, которую будем передавать в знаменатель
# для успешной обработки нулевых значений
a = 0.00001

def smape(target, predictions):
    smape = 1 / len(target) * sum (abs(target - predictions) / 
                                  ((abs(target) + abs(predictions)) / 2)) * 100
    return smape



def final_smape(smape_raugher, smape_final):
    final_smape = smape_raugher * 0.25 + smape_final * 0.75
    return final_smape

smape_score = make_scorer(smape, greater_is_better= False )
```
